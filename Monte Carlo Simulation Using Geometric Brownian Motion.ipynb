## Monte Carlo Simulation: Overview

Monte Carlo simulation is a statistical technique that models the probability of different outcomes in a process that cannot easily be predicted due to the intervention of random variables. It is used to understand the impact of risk and uncertainty in forecasting and decision-making. The simulation runs a large number of scenarios, known as "iterations," based on random inputs, to generate a range of possible outcomes.

### Uses and Applications of Monte Carlo Simulation:

1. **Finance**:
   - **Stock Price Prediction**: Monte Carlo simulations model the future movement of stock prices by accounting for randomness (e.g., market conditions, investor sentiment).
   - **Portfolio Management**: Investors use it to assess the potential risk and return of a portfolio, helping them make better investment decisions.
   - **Option Pricing**: Monte Carlo methods are used to price complex financial derivatives like European and American options, especially when closed-form solutions are unavailable.
   - **Risk Management**: It helps quantify financial risks, such as market or credit risk, and estimate potential losses under different scenarios.

2. **Business & Operations**:
   - **Project Management**: Monte Carlo simulation is used to estimate the probability of completing a project within a certain timeline and budget, given the uncertainties involved in the process.
   - **Supply Chain Management**: Businesses simulate supply chain disruptions, forecast inventory requirements, and optimize order policies.

3. **Engineering & Manufacturing**:
   - **Quality Control**: Engineers use Monte Carlo simulations to test product designs by modeling the effect of variations in materials, manufacturing processes, and environmental conditions.
   - **System Reliability**: It helps to evaluate the reliability of complex systems by simulating failure scenarios.

4. **Science & Medicine**:
   - **Clinical Trials**: Researchers simulate patient outcomes in clinical trials to predict the efficacy of a new treatment, accounting for variability in patient responses.
   - **Physics & Biology**: Monte Carlo methods are employed in simulations of physical and biological systems, such as molecular dynamics, radiation therapy planning, and statistical physics.

### Basic Info on Monte Carlo Simulation:

Monte Carlo simulation generates random inputs based on specified probability distributions (e.g., normal distribution for stock prices). These inputs are then used to compute the possible outcomes of a mathematical model. By repeating this process thousands of times, the simulation provides a distribution of potential outcomes, which can be used for decision-making.

### Why Monte Carlo Simulation?
- **Handles Uncertainty**: Many real-world problems involve unpredictable variables, and Monte Carlo simulations allow us to model this randomness.
- **Insight into Risk**: By generating thousands of possible scenarios, Monte Carlo provides a better understanding of potential risks and the likelihood of different outcomes.
- **Versatility**: It can be applied across various domains, including finance, business, engineering, and science.


## Formula Used in Monte Carlo Simulation for Stock Price Prediction

In finance, stock prices are often modeled using **Geometric Brownian Motion (GBM)**. This model captures two key aspects of stock prices:
1. **Drift**: The deterministic upward or downward trend, representing the expected return.
2. **Volatility**: The random component that introduces uncertainty or risk in the stock price.

The general formula for the stock price \( S_t \) at time \( t \) using GBM is:

$$
S_t = S_0 \times \exp\left[\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right]
$$

Where:
- $S_t$ is the stock price at time \( t \).
- $S_0$ is the initial stock price at time 0.
- $u$ is the expected return (drift rate).
- $sigma$ is the volatility (standard deviation of returns).
- $W_t$ is the Wiener process (random term representing uncertainty, drawn from a normal distribution).
- $t$ is the time period over which the simulation is run.

### Breakdown of the Formula:

1. **Drift Term**
   $$
    \left(\mu - \frac{\sigma^2}{2}\right)
   $$
   - This part accounts for the **deterministic growth** of the stock price.
   - $u$ is the expected return (growth rate of the stock price) over time.
   - The term $\frac{\sigma^2}{2} $ adjusts for volatility, preventing overestimation of the stock's future value. Without this term, the model would over-predict the expected future value.

3. **Volatility Term**
    $$ \sigma W_t $$
   - This part captures the **randomness** or **uncertainty** in stock prices, making the model stochastic (random).
   - $ W_t $ represents the **Wiener process** or Brownian motion, which generates random numbers from a normal distribution. It models the uncertainty in stock price movements due to various unpredictable factors.
   - $ \sigma $ controls the amount of variation expected from the stock’s mean return. A higher $ \sigma $ means more uncertainty.

5. **Exponential Term**
   $$ \exp $$
   - The exponential function ensures that the stock price remains positive, as stock prices cannot be negative. It converts the combined effect of drift and volatility into multiplicative changes in price over time.

### Stochastic Nature of the Model:
The **stochastic** part of the formula $ W_t $ means that every time we run the Monte Carlo simulation, different outcomes are generated based on randomly selected values for $ W_t $. By repeating this process many times, we can build a distribution of possible stock prices at future points in time.


## How the Formula Works

In this section, we will break down how the formula used in Monte Carlo simulation for stock price prediction works in practice. The formula is based on **Geometric Brownian Motion (GBM)** and incorporates two main components: **drift** and **volatility**.

The formula is:

$$
S_t = S_0 \times \exp\left[\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right]
$$

### Step-by-Step Explanation:

### 1. **Initial Price** $S_0$:
   - This is the known starting price of the stock at the beginning of the simulation (at time $t = 0$).
   - Every simulation begins from this initial value, and it will evolve over time based on the following components.

### 2. **Drift Term** $\left(\mu - \frac{\sigma^2}{2}\right)$:
   - $\mu$ is the expected return of the stock, representing the average rate of growth or decline. A positive $\mu$ suggests the stock is expected to increase in price over time.
   - The adjustment $\frac{\sigma^2}{2}$ accounts for volatility. Without this adjustment, the stock price would be consistently overestimated in the presence of high volatility.
   - The drift term defines the **long-term trend** of the stock price without considering randomness. It can be thought of as the deterministic part of the stock price movement.

### 3. **Volatility Term** $\sigma W_t$:
   - This term represents the **random variation** of the stock price due to unpredictable factors (market movements, investor behavior, news, etc.).
   - $W_t$ is a Wiener process (Brownian motion), which generates random numbers from a normal distribution. These random numbers introduce uncertainty, making each simulation unique.
   - The $\sigma$ (volatility) scales the randomness. A higher $\sigma$ increases the magnitude of possible price changes, leading to larger swings in price. This is the stochastic component of the model, which makes the stock price unpredictable.

### 4. **Exponential Function** $\exp$:
   - The use of the exponential function ensures that stock prices remain positive, as stock prices cannot go below zero.
   - The function applies the combined effect of the drift and volatility terms multiplicatively, generating a compounded effect on the stock price over time.

### 5. **Time Increment** $t$:
   - The time $t$ defines how far into the future the simulation is being run. If we want to simulate a stock’s behavior over 1 year, we divide the year into small time increments (e.g., days or months).
   - As we increase $t$, the effect of both the drift and the random volatility accumulates, resulting in a possible range of outcomes for the stock price.

### 6. **Multiple Simulations**:
   - Since Monte Carlo simulation is a probabilistic model, we repeat the process multiple times (with thousands of iterations) to get a distribution of possible future stock prices.
   - Each simulation gives a different outcome due to the random nature of the volatility term $W_t$. By averaging across these simulations, we can estimate the expected future price as well as the range of possible outcomes.

### Summary:
- The formula starts from an initial stock price $S_0$.
- The drift term models the expected average behavior (e.g., growth).
- The volatility term introduces randomness, simulating the unpredictable changes in price.
- The exponential function ensures the prices evolve multiplicatively and stay positive.
- By iterating the formula over many time steps and simulations, we generate a range of possible stock prices at a future date.

This combination of deterministic and random components captures both the **expected trend** and **uncertainty** of the stock price, making Monte Carlo simulation a powerful tool for predicting future price movements.


## Parameters

In this section, we will describe the parameters required for the Monte Carlo simulation of stock price prediction. Each of these parameters is essential for running the simulation and directly influences the outcome. Here are the key parameters we will use:

### 1. **Initial Stock Price ($S_0$)**:
   - This is the starting price of the stock at the beginning of the simulation.
   - We typically obtain this value from the most recent historical stock data.

### 2. **Expected Return ($\mu$)**:
   - The expected rate of return for the stock, which can be derived from historical data or estimated based on future expectations.
   - In practice, we may calculate $\mu$ as the mean of daily returns over a specified historical period.

### 3. **Volatility ($\sigma$)**:
   - The measure of how much the stock price fluctuates, often represented as the standard deviation of stock returns.
   - Like $\mu$, we calculate volatility using historical stock data. It reflects the stock's risk level — higher volatility means more significant price fluctuations.

### 4. **Time Horizon ($T$)**:
   - The period over which we simulate the stock price. For example, if we want to simulate the price over 1 year, we set $T = 1$.
   - We break this time period into smaller intervals (e.g., daily intervals), depending on the level of granularity desired.

### 5. **Number of Time Steps**:
   - This is the number of subdivisions of the time horizon $T$. For instance, if we simulate the stock price over 1 year, we could use 252 time steps to represent each trading day in a year.
   - Smaller time steps produce more detailed simulations but require more computational power.

### 6. **Number of Simulations**:
   - The Monte Carlo method is a probabilistic approach, so to get a good estimate of potential future stock prices, we need to repeat the simulation many times.
   - The number of simulations determines how many possible price paths we generate. A higher number of simulations gives a broader distribution of possible outcomes.

### Summary of Parameters:
- **$S_0$**: Initial stock price.
- **$\mu$**: Expected return (mean of historical returns).
- **$\sigma$**: Volatility (standard deviation of returns).
- **$T$**: Time horizon (e.g., 1 year).
- **Number of time steps**: Subdivisions of the time horizon (e.g., 252 days).
- **Number of simulations**: Total iterations of the simulation to produce a range of outcomes.


## 1. Getting the Data from Historical Prices

To run the Monte Carlo simulation, we need historical stock price data to estimate key parameters like **expected return** ($\mu$) and **volatility** ($\sigma$). This data can be obtained using APIs or libraries such as `yfinance` or `pandas_datareader`.

### Steps to Get Historical Data:

1. **Use `yfinance` or other data sources**:
   - We will use the `yfinance` package to download historical stock prices. This package provides daily data on open, high, low, close, and volume, which are crucial for calculating returns and volatility.

2. **Calculate Daily Returns**:
   - From the historical adjusted closing prices, we calculate daily returns using the formula:

   \[
   R_t = \frac{P_t}{P_{t-1}} - 1
   \]

   where $R_t$ is the return at time $t$, and $P_t$ is the adjusted closing price at time $t$.

3. **Calculate Expected Return ($\mu$)**:
   - The mean of the daily returns is used to compute the expected return. We then annualize this value by multiplying it by the number of trading days in a year (usually 252 days).

4. **Calculate Volatility ($\sigma$)**:
   - The standard deviation of the daily returns gives the volatility. Similar to the expected return, we annualize this value by multiplying it by the square root of 252.

### Code to Get Historical Data

Here’s a Python code snippet using `yfinance` to retrieve historical stock data and calc: {mu}")
print(f"Volatility (sigma): {sigma}")


```python
import yfinance as yf
import numpy as np

# Define the stock symbol and download historical data
stock_symbol = 'NVDA'
data = yf.download(stock_symbol, start='2020-01-01', end='2024-01-01')

# Calculate daily returns
data['Returns'] = data['Adj Close'].pct_change()

# Calculate expected return (mean of daily returns) and annualize it
mu = data['Returns'].mean() * 252

# Calculate volatility (standard deviation of daily returns) and annualize it
sigma = data['Returns'].std() * np.sqrt(252)

# Display results
print(f"Expected return (mu): {mu}")
print(f"Volatility (sigma): {sigma}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = data['Adj Close'][-1]  # Initial stock price (most recent price)
T = 1  # Time horizon (1 year)
mu = data['Returns'].mean() * 252  # Expected annual return
sigma = data['Returns'].std() * np.sqrt(252)  # Annualized volatility
n_simulations = 1000  # Number of simulations
n_days = 252  # Number of trading days in a year

# Time increments (252 trading days over 1 year)
dt = T / n_days

# Simulate random values for the Wiener process
Z = np.random.standard_normal((n_days, n_simulations))

# Create empty array to hold the simulated stock prices
simulated_prices = np.zeros((n_days, n_simulations))
simulated_prices[0] = S0  # All simulations start from S0

# Monte Carlo simulation loop
for t in range(1, n_days):
    simulated_prices[t] = simulated_prices[t - 1] * np.exp(
        (mu - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z[t]
    )

# Plot the simulated paths
plt.figure(figsize=(10, 6))
plt.plot(simulated_prices)
plt.title(f"Monte Carlo Simulation of {n_simulations} Stock Price Paths (1 Year)")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```

### Explanation:

- **S0**: The initial stock price (most recent adjusted close price).
- **T**: Time horizon (1 year).
- **mu**: Expected annual return, calculated as the mean of daily returns multiplied by 252 (the number of trading days in a year).
- **sigma**: Volatility, represented by the standard deviation of daily returns and annualized by multiplying by $\sqrt{252}$.
- **n_simulations**: The number of simulated price paths (1000 simulations in this example).
- **n_days**: The number of trading days (252, typically for one year).
- **Z**: Random values from a standard normal distribution used to simulate the Wiener process ($W_t$) in the Monte Carlo simulation.
- **simulated_prices**: A matrix that stores the simulated stock prices for each time step. The first row contains the initial stock price $S_0$, and each subsequent row contains stock prices generated by the Monte Carlo method using the formula:

\[
S_t = S_0 \times \exp\left[\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right]
\]


### VaR and Plotting the Distribution:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the daily returns from the simulated prices
simulated_returns = simulated_prices[-1] / simulated_prices[0] - 1

# Plot the distribution of returns
plt.figure(figsize=(10, 6))
sns.histplot(simulated_returns, bins=50, kde=True, color='blue')
plt.axvline(np.percentile(simulated_returns, 5), color='red', linestyle='dashed', linewidth=2, label="5% VaR")
plt.title('Distribution of Simulated Stock Returns')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the 5% shortfall risk (Value at Risk - VaR)
VaR_5_percent = np.percentile(simulated_returns, 5)
print(f"5% Value at Risk (VaR): {VaR_5_percent:.2%}")
```

## Interpretation of Monte Carlo Simulation and Value at Risk (VaR) Outputs

**Key Takeaways**:
